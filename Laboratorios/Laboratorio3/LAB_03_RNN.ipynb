# **MODELO RNN CON DATOS HISTORICOS SOBRE LA CALIDAD DEL AGUA**

## INVESTIGACIÓN SOBRE LOS INDICADORES DE LA CALIDAD DEL AGUA


Para realizar una predicción efectiva del índice de calidad del agua utilizando una Red Neuronal Recurrente (RNN), es crucial identificar los indicadores clave que determinan la calidad del agua. Estos indicadores se basan en parámetros físico-químicos que evalúan el estado del agua en términos de contaminación, salud pública y capacidad para sostener la vida acuática. A continuación, explicaremos los indicadores más relevantes que se deben considerar, basándome en fuentes estándar como la Organización Mundial de la Salud (OMS) y el Índice de Calidad del Agua (ICA).

### Indicadores físico-químicos a considerar:

### 1. pH:

- Descripción: El pH mide la acidez o alcalinidad del agua.
- Límites: Un pH adecuado para agua potable y ecosistemas acuáticos debe estar en el rango de 6.5 a 8.5.
- Importancia: El pH fuera de este rango puede ser indicativo de contaminación química o biológica y puede afectar la vida acuática.

### 2. Oxígeno Disuelto (OD):

- Descripción: Es una medida de la cantidad de oxígeno disponible en el agua para la respiración de organismos acuáticos.
- Límites: Niveles de oxígeno disuelto por encima de 6 mg/L son ideales para mantener ecosistemas saludables.
- Importancia: El oxígeno disuelto es esencial para la vida acuática. Bajos niveles pueden indicar contaminación orgánica o excesiva demanda biológica.

### 3. Demanda Bioquímica de Oxígeno (DBO):

- Descripción: Mide la cantidad de oxígeno requerido por organismos para descomponer la materia orgánica en el agua.
- Límites: Valores bajos de DBO (menos de 5 mg/L) indican buena calidad de agua. Niveles altos indican contaminación orgánica.
- Importancia: Un aumento en la DBO indica la presencia de contaminantes orgánicos que consumen el oxígeno, afectando la vida acuática.

### 4. Sólidos Disueltos Totales (TDS):

- Descripción: Mide la concentración de partículas disueltas en el agua, como minerales, sales y metales.
- Límites: El límite para agua potable es de aproximadamente 500 mg/L.
- Importancia: Niveles altos de TDS afectan la calidad del agua para consumo humano y uso agrícola.

### 5. Conductividad Eléctrica (CE):

- Descripción: La conductividad mide la capacidad del agua para conducir electricidad, relacionada con la concentración de sales disueltas.
- Límites: Valores adecuados son generalmente entre 50 y 1500 µS/cm.
- Importancia: Es un indicador de la presencia de iones disueltos en el agua, que puede afectar tanto la potabilidad como los ecosistemas.

### 6. Turbidez:

- Descripción: Mide la claridad del agua, que puede ser afectada por partículas suspendidas.
- Límites: Agua potable debe tener una turbidez menor a 1 NTU.
- Importancia: Alta turbidez puede significar contaminación física, como sedimentos o presencia de microorganismos patógenos.

### 7. Nitratos (NO3-):

- Descripción: La concentración de nitratos proviene de la descomposición de materia orgánica y el uso de fertilizantes.
- Límites: El valor límite para agua potable es de 50 mg/L.
- Importancia: Altas concentraciones pueden causar eutrofización y afectar la salud humana, causando trastornos como el síndrome del bebé azul.

### 8. Fosfatos (PO4):

- Descripción: Los fosfatos provienen principalmente de detergentes y fertilizantes.
- Límites: Valores bajos de fosfatos, típicamente menos de 0.1 mg/L, son deseables para evitar la proliferación de algas.
- Importancia: Exceso de fosfatos puede provocar floración algal y eutrofización, afectando gravemente la calidad del agua.

### 9. Amonio (NH4+):

- Descripción: El amonio es una forma de nitrógeno que proviene de residuos orgánicos y aguas residuales.
- Límites: Para aguas de consumo, los niveles deben ser inferiores a 0.5 mg/L.
- Importancia: Indica contaminación por desechos orgánicos. Niveles elevados pueden ser tóxicos para la vida acuática.


### 10. Cloruros (Cl-):

- Descripción: Los cloruros son sales disueltas que pueden provenir de aguas residuales o contaminación industrial.
- Límites: El nivel recomendado es de 250 mg/L.
- Importancia: Altos niveles pueden afectar el sabor del agua y ser perjudiciales para ciertos cultivos.

### 11. Metales Pesados (Plomo, Mercurio, Cadmio, etc.):

- Descripción: Estos metales son contaminantes tóxicos que pueden acumularse en el cuerpo y causar problemas de salud graves.
- Límites: Cada metal tiene su propio límite; por ejemplo, el plomo tiene un límite de 0.01 mg/L.
- Importancia: Los metales pesados son tóxicos, incluso en bajas concentraciones, y pueden contaminar fuentes de agua por actividades industriales.

### 12. Coliformes Fecales:

- Descripción: Indican la posible presencia de microorganismos patógenos, como bacterias fecales.
- Límites: Para agua potable, debe estar libre de coliformes.
- Importancia: La presencia de coliformes fecales es un indicador de contaminación biológica y riesgo para la salud humana.

## 2. Definir el Dataset para el modelo LSTM

```python
import torch
from torch.utils.data import Dataset, DataLoader

class WaterQualityDataset(Dataset):
    def __init__(self, X, y):
        self.X = X
        self.y = y

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        # Expandir la dimensión de los datos de entrada para agregar el time_step si es necesario
        X_data = torch.tensor(self.X[idx], dtype=torch.float32).unsqueeze(0)  # Agregar dimensión temporal
        y_data = torch.tensor(self.y[idx], dtype=torch.float32)
        return X_data, y_data
```

```python
# Crear DataLoaders para entrenamiento, validación y prueba

train_dataset = WaterQualityDataset(X_train_scaled, y_train)
valid_dataset = WaterQualityDataset(X_valid_scaled, y_valid)
test_dataset = WaterQualityDataset(X_test_scaled, y_test)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
valid_loader = DataLoader(valid_dataset, batch_size=64, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)
```

## 3. Definir el Modelo LSTM

```python
import torch.nn as nn

class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size, dropout=0.2):
        super(LSTMModel, self).__init__()
        self.lstm = nn.LSTM(input_size=input_size, hidden_size=hidden_size, num_layers=num_layers, batch_first=True, dropout=dropout)
        self.fc = nn.Linear(hidden_size, output_size)


    def forward(self, x):
        if x.dim() == 2:  # Si tiene solo (batch_size, num_features)
            x = x.unsqueeze(1)  # Añade la dimensión temporal (batch_size, 1, num_features)

        lstm_out, _ = self.lstm(x)  # Salida de la LSTM, debería tener la forma (batch_size, time_steps, hidden_size)

        # Verifica las dimensiones antes de indexar
        if lstm_out.dim() == 3:
            out = self.fc(lstm_out[:, -1, :])  # Usar solo la salida del último paso temporal
        else:
            raise ValueError(f"Unexpected LSTM output shape: {lstm_out.shape}")

        return out
```

```python
# Parámetros del modelo
input_size = X_train_scaled.shape[1]  # Número de características (pH, Iron, etc.)
hidden_size = 64  # Tamaño de las capas ocultas
num_layers = 2  # Número de capas LSTM
output_size = 1  # Predicción de la calidad del agua

# Instanciar el modelo
model = LSTMModel(input_size, hidden_size, num_layers, output_size)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
```

## 4. Entrenamiento del Modelo

```python
import os
import torch.optim as optim
from tqdm import tqdm  # Importar tqdm para la barra de progreso

def fit(model, train_loader, valid_loader, epochs=50, lr=0.3, patience=5, save_dir='checkpoints/'):
    optimizer = optim.Adam(model.parameters(), lr=lr)
    criterion = nn.MSELoss()
    best_loss = float('inf')
    patience_counter = 0

    # Crear la carpeta si no existe
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)

    train_losses, valid_losses = [], []

    for epoch in range(epochs):
        model.train()
        running_loss = 0.0

        # Crear barra de progreso para el entrenamiento
        progress_bar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}", leave=False)

        for X_batch, y_batch in progress_bar:
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)
            optimizer.zero_grad()
            y_pred = model(X_batch)
            loss = criterion(y_pred, y_batch.unsqueeze(1))
            loss.backward()
            optimizer.step()
            running_loss += loss.item()

            # Actualizar la barra de progreso con la pérdida
            progress_bar.set_postfix(train_loss=loss.item())

        train_loss = running_loss / len(train_loader)
        train_losses.append(train_loss)

        # Validación
        model.eval()
        valid_loss = 0.0
        with torch.no_grad():
            for X_batch, y_batch in valid_loader:
                X_batch, y_batch = X_batch.to(device), y_batch.to(device)
                y_pred = model(X_batch)
                loss = criterion(y_pred, y_batch.unsqueeze(1))
                valid_loss += loss.item()

        valid_loss = valid_loss / len(valid_loader)
        valid_losses.append(valid_loss)

        # Mostrar las pérdidas de entrenamiento y validación para cada época
        print(f"Epoch [{epoch+1}/{epochs}], Train Loss: {train_loss:.4f}, Valid Loss: {valid_loss:.4f}")

        # Early Stopping y guardado del mejor modelo
        if valid_loss < best_loss:
            best_loss = valid_loss
            model_path = os.path.join(save_dir, f'best_lstm_model_epoch{epoch+1}.pth')
            torch.save(model.state_dict(), model_path)  # Guardar el modelo en la carpeta especificada
            patience_counter = 0
        else:
            patience_counter += 1
            if patience_counter >= patience:
                print("Early stopping triggered!")
                break

    return train_losses, valid_losses
```

```python
# Entrenar el modelo
epochs = 50
train_losses, valid_losses = fit(model, train_loader, valid_loader, epochs=50, save_dir='/content/drive/MyDrive/SIS420-IA2/Laboratorios/LAB-03/')
```

## 5. Evaluación y Predicciones

Después del entrenamiento, se evaluará el modelo en el conjunto de prueba y se harán las predicciones necesarias para 2024 y 2025.

```python
import torch
from torch.utils.data import Dataset, DataLoader

class WaterQualityDataset(Dataset):
    def __init__(self, X, y):
        self.X = X
        self.y = y

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        X_data = torch.tensor(self.X[idx], dtype=torch.float32).unsqueeze(0)  # Añadimos la dimensión temporal
        y_data = torch.tensor(self.y[idx], dtype=torch.float32)
        return X_data, y_data

# Crear DataLoaders
train_dataset = WaterQualityDataset(X_train_scaled, y_train)
valid_dataset = WaterQualityDataset(X_valid_scaled, y_valid)
test_dataset = WaterQualityDataset(X_test_scaled, y_test)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
valid_loader = DataLoader(valid_dataset, batch_size=64, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)
```

## 1. Definir el modelo GRU

```python
import torch.nn as nn

class GRUModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size, dropout=0.3):
        super(GRUModel, self).__init__()
        self.gru = nn.GRU(input_size=input_size, hidden_size=hidden_size, num_layers=num_layers, batch_first=True, dropout=dropout)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        if x.dim() == 2:  # Si tiene solo (batch_size, num_features)
            x = x.unsqueeze(1)  # Añade la dimensión temporal (batch_size, 1, num_features)

        gru_out, _ = self.gru(x)  # Salida de la GRU, forma (batch_size, time_steps, hidden_size)

        # Usar solo la salida del último paso temporal
        out = self.fc(gru_out[:, -1, :])
        return out
```

## 2. Configurar el entrenamiento

Usamos el optimizador Adam y la función de pérdida MSE. También implementamos el early stopping para evitar el sobreajuste.

```python
import os
import torch.optim as optim
from tqdm import tqdm  # Barra de progreso

def fit(model, train_loader, valid_loader, epochs=50, lr=0.001, patience=7, save_dir='checkpoints/'):
    optimizer = optim.Adam(model.parameters(), lr=lr)
    criterion = nn.MSELoss()
    best_loss = float('inf')
    patience_counter = 0

    if not os.path.exists(save_dir):
        os.makedirs(save_dir)

    train_losses, valid_losses = [], []

    for epoch in range(epochs):
        model.train()
        running_loss = 0.0
        progress_bar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}", leave=False)

        for X_batch, y_batch in progress_bar:
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)
            optimizer.zero_grad()
            y_pred = model(X_batch)
            loss = criterion(y_pred, y_batch.unsqueeze(1))
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
            progress_bar.set_postfix(train_loss=loss.item())

        train_loss = running_loss / len(train_loader)
        train_losses.append(train_loss)

        # Validación
        model.eval()
        valid_loss = 0.0
        with torch.no_grad():
            for X_batch, y_batch in valid_loader:
                X_batch, y_batch = X_batch.to(device), y_batch.to(device)
                y_pred = model(X_batch)
                loss = criterion(y_pred, y_batch.unsqueeze(1))
                valid_loss += loss.item()

        valid_loss = valid_loss / len(valid_loader)
        valid_losses.append(valid_loss)

        print(f"Epoch [{epoch+1}/{epochs}], Train Loss: {train_loss:.4f}, Valid Loss: {valid_loss:.4f}")

        # Early Stopping y guardar el mejor modelo
        if valid_loss < best_loss:
            best_loss = valid_loss
            torch.save(model.state_dict(), os.path.join(save_dir, f'best_gru_model_epoch{epoch+1}.pth'))
            patience_counter = 0
        else:
            patience_counter += 1
            if patience_counter >= patience:
                print("Early stopping triggered!")
                break

    return train_losses, valid_losses
```
